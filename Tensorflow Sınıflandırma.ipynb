## Sınıflandırma Modeli Oluşturma

```python
y = df["Type"].values
x= df.drop("Type", axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train = scaler.transform(x_train)
```

```python
x_test = scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation, Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#earlystopping ve dropout u overfitting problemleri için kullanıcaz
```

```python
x_train.shape
```

```python
model = Sequential()
model.add(Dense(units=30, activation="relu"))  #içeriye 30 tane nöron koy. (genelgeçer bir kural, yapıp deniyoruz)
model.add(Dense(units=15, activation="relu")) 
model.add(Dense(units=15, activation="relu")) 
model.add(Dense(units=1, activation="sigmoid")) #sigmoid 0-1 arasında bir değer veriyor.

model.compile(loss="binary_crossentropy", optimizer="adam") #loss hep mse veriyorduk ama bu sefer sınıflandırma işlemi yapıyoruz binary olduğu için bunu verdik.
#neden 30? -->> kaç tane column varsa giriş layer'ını o kadar yapmamız önerilir. üst kodda gördüğümüz üzere 30 column var.
#son layer'a 1 koyuyoruz.
#aradaki layerlara 30 ve 1 arasında bir sayı seçip koymamız önerilir
```

```python
model.fit(x_train, y_train, epochs=700, validation_data=(x_test, y_test), verbose=1) #epochs yüksek verip deneyelim. overfittinge yol açacaktır muhtemelen
```

verbose=1 Ne Demek?


verbose=1, eğitim sürecinde konsola eğitim aşamalarıyla ilgili bilgi verilmesini sağlar.

*verbose=0: Eğitim süreci sırasında hiçbir çıktı göstermez.

*verbose=1: Eğitim sırasında her epoch için eğitim ve doğrulama kayıplarını ve metriklerini adım adım ekrana yazdırır.

*verbose=2: Her epoch’un sadece özet bilgisini ekrana yazar (daha az detaylıdır).


verbose=1, özellikle model eğitimi sırasında her bir epoch’ta ne kadar ilerleme kaydedildiğini görmek için kullanılı

```python
modelKaybi.plot()
#val_loss, loss ile beraber aşağı doğru gitmesi gerekirken yükselmiş.
```

```python
model = Sequential()
model.add(Dense(units=30, activation="relu"))  #içeriye 30 tane nöron koy. (genelgeçer bir kural, yapıp deniyoruz)
model.add(Dense(units=15, activation="relu")) 
model.add(Dense(units=15, activation="relu")) 
model.add(Dense(units=1, activation="sigmoid")) #sigmoid 0-1 arasında bir değer veriyor.

model.compile(loss="binary_crossentropy", optimizer="adam") #loss hep mse veriyorduk ama bu sefer sınıflandırma işlemi yapıyoruz binary olduğu için bunu verdik.
#neden 30? -->> kaç tane column varsa giriş layer'ını o kadar yapmamız önerilir. üst kodda gördüğümüz üzere 30 column var.
#son layer'a 1 koyuyoruz.
#aradaki layerlara 30 ve 1 arasında bir sayı seçip koymamız önerilir
```

```python
earlyStopping = EarlyStopping(monitor = "val_loss", mode = "min", verbose=1, patience=25)
#EarlyStopping üstüne shift tab yapıp incele.
#val_loss u en azda tutmaya çalışıyoruz o yüzden mode=min, başka bir şeye bakıyor olsaydık(mesela accuracy) mode u değiştirebilirdik.
```

```python
model.fit(x_train, y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[earlyStopping])
```

```python
modelKaybi = pd.DataFrame(model.history.history)
modelKaybi.plot()
#gözle görülür bir iyileştirme var. peki hala memnun değilsek ne yapacağız?
#dropout tekniğini kullanacağız
```

Dropout, derin öğrenme modellerinde aşırı öğrenmeyi (overfitting) engellemek için kullanılan popüler bir düzenleme (regularization) tekniğidir.


Dropout, eğitim sırasında rastgele seçilen nöronları geçici olarak devre dışı bırakma yöntemidir. Bu, nöronların modelin o eğitim adımında katkıda bulunmamasını sağlar. Böylece, model her eğitim adımında farklı bir alt ağ öğrenir ve bu sayede modelin aşırı öğrenme yapması zorlaşır. Sonuç olarak, model daha sağlam ve genellemeye açık hale gelir.



Nasıl Çalışır?
Eğitim sırasında, her adımda belirli bir oran (örneğin %20 veya %50) kadar nöron rastgele seçilerek devre dışı bırakılır.
Bu nöronlar o adımda çıkış üretmez, yani öğrenmeye dahil olmazlar.
Bu işlem sadece eğitim sırasında uygulanır. Test sırasında modelin tüm nöronları devrede olur.


Dropout, layerlarla ilgili bir overfitting varsa o da training esnasında daha iyi hale gelecek, val_lossu daha kontrol altına alacak

```python
model = Sequential()

model.add(Dense(units=30, activation="relu"))  
model.add(Dropout(0.4))

model.add(Dense(units=15, activation="relu"))
model.add(Dropout(0.4))

model.add(Dense(units=15, activation="relu")) 
model.add(Dropout(0.4))

model.add(Dense(units=1, activation="sigmoid")) 

model.compile(loss="binary_crossentropy", optimizer="adam") 
#dropout rate 0 ile 50 arasında seçilmeli. 0.6 yüksek bir oran olurdu ve hatta normalde küçük bir uyarı verir Large Dropout Rate diye.
```

```python
model.fit(x=x_train, y=y_train, epochs=700, validation_data=(x_test, y_test), verbose=1, callbacks=[earlyStopping])
```

```python
kayipDF = pd.DataFrame(model.history.history)
kayipDF.plot()
#dropout kullanınca losslar zikzaklı olabilir.
```

```python
tahminler_prob = model.predict(x_test)
print(tahminler_prob)
tahminler = np.argmax(tahminler_prob, axis=1)  # Sınıf etiketlerini elde et
tahminler = (tahminler_prob > 0.5).astype(int)  # Olasılık 0.5'ten büyükse 1, küçükse 0
print(tahminler)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test, tahminler))
```

```python
confusion_matrix(y_test, tahminler)
```
